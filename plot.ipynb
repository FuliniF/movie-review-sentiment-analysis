# For figures in experiments

```python
models = ['Naive Bayes', 'SVM', 'Lbl2Vec']
categories = ['Accuracy', 'Presicion', 'Recall', 'F1', 'AUROC']
color = ['lightseagreen', 'lightyellow', 'lightsalmon']
# color = ['lightpink', 'lightblue', 'lightseagreen', 'lightyellow', 'lightsalmon']
```

```python

data = {
    'Naive Bayes': [79.8488120950324, 93.83730484798686, 74.47016628627323, 83.03944737320488, 82.43661865177385],
    'SVM': [86.76291528961558, 87.32344632768362, 86.01739130434782, 86.66549851060102, 86.7630449910744],
    'Lbl2Vec': [73.06695464362851, 79.60312296681847, 79.7848059993479, 79.69386093470119, 69.83482142577759]
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

bar_width = 0.25
index = np.arange(len(categories))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for i, model in enumerate(models):
    ax.bar(index + i*bar_width, data[model], bar_width, label=model, color=color[i], edgecolor='black')

ax.set_xticks(index + bar_width)
ax.set_xticklabels(categories)

ax.set_ylim(40,100)

ax.set_xlabel('Metrics')
ax.set_ylabel('Score (%)')
ax.set_title('Testing Performance with Balanced Training Dataset')
ax.legend()

plt.show()
```

```python

data = {
    'Imbalanced': [74.23326133909287, 88.18255908720456, 70.55754809259863, 78.39159572541206, 76.00174269633131],
    'Balanced': [73.06695464362851, 79.60312296681847, 79.7848059993479, 79.69386093470119, 69.83482142577759]
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

bar_width = 0.35
index = np.arange(len(categories))
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.bar(index + bar_width/2, data['Imbalanced'], bar_width, label='Imbalanced', color='lightsalmon', edgecolor='black')
ax.bar(index + bar_width/2 *3, data['Balanced'], bar_width, label='Balanced', color='lightsalmon', edgecolor='black')

ax.set_xticks(index + bar_width)
ax.set_xticklabels(categories)

ax.set_ylim(40,100)

ax.set_xlabel('Metrics')
ax.set_ylabel('Score (%)')
ax.set_title('Testing Performance on Lbl2Vec')
ax.legend()

plt.show()
```
